### Running SCOOTR on CITE-seq dataset:

```python
from scootrv2 import *

# Read in CITE-seq data. ADT = Antibodies, RNA= Gene expression:
adt=np.genfromtxt("../data/citeseq_adt_normalized_1000cells.csv", delimiter=",")
rna=unit_normalize(np.genfromtxt("../data/citeseq_rna_normalizedFC_1000cells.csv", delimiter=","))

print(adt.shape, rna.shape)
```

```python
#These datasets are of form [n_feat x n_samples] but the algorithm expects [n_samples x n_feat]
#So we transpose them below. Also, we found normalization useful for all OT applications on single-cell datasets (including SCOT and SCOTv2):
adt=unit_normalize(np.transpose(adt))
rna=unit_normalize(np.transpose(rna))
```

```python
# Compute alignments:
Ts, Tv, cost, log_out= scootr(adt,rna, alpha=0.5, k=25, reg=0.001,reg2=0.1,verbose=True)
```

Here in the output, Ts and Tv give us the coupling (correspondence/alignment) matrices:<br>
**Ts**= correspondence matrix between the SAMPLES<br>
**Tv**= correspondence matrix between the FEATURES<br>

**Additional notes on the function hyperparameters:**<br>
**reg**= coefficient of entropic regularization (epsilon) for SAMPLE alignments<br>
**reg2**= coefficient of entropic regularization (epsilon) for FEATURE alignments<br>
**alpha**= interpolation coefficient between the graph distances and euclidean distances between samples. Default=0.5. Higher values of alpha values graph distances more than COOT distance and vice-versa<br>
**k**= number of neighbors to use in kNN graph when computing graph distances for samples. Default=25 <br>



### Evaluating the alignments

```python
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA #For visualization of the cell-cell alignments
from utils import * #contains barycentric projection code
from evals import * #contains FOSCTTM code
```

```python
#Perform barycentric projection based on sammple alignment matrix, Ts:
adt_aligned,rna_aligned=barycentric_projection(adt,rna,Ts)
# Compute mean FOSCTTM (lower the better, ranges between 0-1)
print(np.mean(calc_domainAveraged_FOSCTTM(adt_aligned,rna_aligned)))
```

```python
# Visualize feature alignments
# Note that the features were ordered in the data files so the correct alignment would lie in the diagonal:

# Gene names(on the x axis because antibody features are in the rows of the coupling matrix):
x_labels=["ITGAL","ITGAX","IL3RA","IL7R","CD14","FCGR3A","KLRB1","CD19","CCR7","IL2RA","CD27","ICOS","CD28","CD3E","CD34","CD38","CD4","PTPRC","PTPRC","NCAM1","B3GAT1","CD69","CD79B","CD8A","HLA-DRA"] 
# Antibody names (on the y axis because antibody features are in the rows of the coupling matrix):
y_labels=["CD11a","CD11c","CD123","CD127-IL7Ra","CD14","CD16","CD161","CD19","CD197-CCR7","CD25","CD27","CD278-ICOS","CD28","CD3","CD34","CD38","CD4","CD45RA","CD45RO","CD56","CD57","CD69","CD79b","CD8a","HLA.DR"]


fig, ax = plt.subplots()
ax.imshow(Tv,cmap="Reds")
ax.set_xticks(np.arange(len(x_labels)))
ax.set_yticks(np.arange(len(y_labels)))
ax.set_xticklabels(x_labels,rotation=90)
ax.set_yticklabels(y_labels)
plt.show()
```

```python
#Visualize cell-cell alignments by domain:
pca=PCA(n_components=2)
#We will perfom PCA jointly on the two domains so the PCA factors/axes are the same:
adt_rna_joint=np.concatenate((adt_aligned,rna_aligned),axis=0)
print(adt_rna_joint.shape)

adt_rna_joint_2Dpca=pca.fit_transform(adt_rna_joint)
adt_2Dpca=adt_rna_joint_2Dpca[0:1000,:] #first 1000 samples are from the antibody domain
rna_2Dpca=adt_rna_joint_2Dpca[1000:,:] #next 1000 samples are from the rna domain

plt.scatter(adt_2Dpca[:,0],adt_2Dpca[:,1], c="k", s=10, label="adt")
plt.scatter(rna_2Dpca[:,0],rna_2Dpca[:,1], c="r", s=5, label="rna")
plt.legend()
plt.show()

# NOTE: the plot looks ugly because PCA isn't a great way to cluster cells in this dataset, but TSNE etc looks nicer if you'd like to try.
# NOTE2: I'll later share cell-type annotations, too. -PD.

```
